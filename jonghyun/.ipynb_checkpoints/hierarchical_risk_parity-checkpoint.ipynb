```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

df = pd.read_csv("../data/sample.csv", index_col="Date").dropna()
df.head()
```

```python
corr_mat = df.corr()
distance_corr = np.sqrt(0.5 * (1 - corr_mat))
distance_corr
```

```python
link = linkage(distance_corr, 'single')
Z = pd.DataFrame(link)

fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z)
plt.show()
```

```python
def get_quasi_diag(link):
    link = link.astype(int)
    sort_ix = pd.Series([link[-1,0], link[-1,1]]) 
    num_items = link[-1, 3]
    
    while sort_ix.max() >= num_items:
        sort_ix.index = range(0, sort_ix.shape[0] * 2, 2)
        
        df0 = sort_ix[sort_ix >= num_items]
        
        i = df0.index
        j = df0.values - num_items
        
        sort_ix[i] = link[j, 0]
        
        df0  = pd.Series(link[j, 1], index=i + 1)
        
        sort_ix = sort_ix.append(df0)
        sort_ix = sort_ix.sort_index()
        
        sort_ix.index = range(sort_ix.shape[0])
        
    return sort_ix.tolist()

sort_ix = get_quasi_diag(link)
sort_ix
```

```python
def get_cluster_var(cov, c_items):
    cov_ = cov.iloc[c_items, c_items]
    ivp = 1./ np.diag(cov_)
    ivp /= ivp.sum()
    
    w_ = ivp.reshape(-1,1)
    c_var = np.dot(np.dot(w_.T, cov_), w_)[0,0]
    
    return c_var

def get_rec_bipart(cov, sort_ix):
    w = pd.Series(1, index=sort_ix)
    c_items = [sort_ix]

    while len(c_items) > 0:
        c_items = [i[int(j):int(k)] for i in c_items for j,k in 
                   ((0, len(i)/2),(len(i)/2, len(i))) if len(i) > 1]
        
        for i in range(0, len(c_items), 2):    
            c_items0 = c_items[i]
            c_items1 = c_items[i + 1]
            
            c_var0 = get_cluster_var(cov, c_items0)
            c_var1 = get_cluster_var(cov, c_items1)
            
            alpha = 1 - c_var0 / (c_var0 + c_var1)
            
            w[c_items0] *= alpha
            w[c_items1] *= 1 - alpha
             
    return w
```

```python
cov_mat = df.cov()
weights = get_rec_bipart(cov_mat, sort_ix)
new_index = [df.columns[i] for i in weights.index]
weights.index = new_index

weights
```
