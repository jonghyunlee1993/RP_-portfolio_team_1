```python
import pandas as pd
import FinanceDataReader as fdr
from copy import deepcopy
from tqdm import tqdm

df = fdr.StockListing('KOSPI')
df
```

```python
error_list = []

for i, line in enumerate(tqdm(df.iterrows(), total=df.shape[0])):  
    line = line[-1]
    try:
        single_firm = fdr.DataReader(line.Symbol, '2011-06-01', '2021-06-30')
        single_firm.loc[:, "Name"] = line.Name
        single_firm = single_firm.reset_index().set_index(["Name", "Date"])

        if i == 0:    
            res_df = deepcopy(single_firm)
        else:
            res_df = res_df.append(single_firm)
    except:
        error_list.append(i)

print(f"Error list: {error_list}")
res_df
```
